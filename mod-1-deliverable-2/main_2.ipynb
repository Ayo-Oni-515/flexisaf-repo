Load the csv file

```python
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv("./company_sales_data.csv")
data
```

Exercise 1: Read total profit of all months and show it using a line plot

```python
# line plot
total_profit = data["total_profit"]
plt.plot(total_profit)
plt.grid()
plt.show()
```

Exercise 2: Read data for bathing soap and facewash of all months and display it using the subplot

```python
bathing_soap = data["bathingsoap"]
face_wash = data["facewash"]

# subplots
plt.subplot(1, 2, 1)
plt.plot(bathing_soap)
plt.title("Bathing Soap")
plt.grid()

plt.subplot(1, 2, 2)
plt.plot(face_wash)
plt.title("Face Wash")
plt.grid()

plt.tight_layout()
plt.show()

```
